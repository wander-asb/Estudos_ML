```python
import pandas            as pd
import numpy             as np
import matplotlib.pyplot as plt
import seaborn           as sn
```

```python
path = 'https://raw.githubusercontent.com/alura-cursos/alura-clustering-validation/base-de-dados/CC%20GENERAL.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.drop(columns=['CUST_ID','TENURE'])
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.median())
```

```python
display(df.isnull().sum().sum())
```

```python
from sklearn.preprocessing import Normalizer
values = Normalizer().fit_transform(df.values)
display(values)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters= 5, n_init = 10, max_iter=300)
y_pred = kmeans.fit_predict(values)
```

```python
from sklearn import metrics
labels = kmeans.labels_
silhueta = metrics.silhouette_score(values, labels, metric='euclidean')
print(silhueta)
```
